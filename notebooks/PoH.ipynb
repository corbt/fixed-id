```python
import pandas as pd

df = pd.DataFrame(logs)

df
```

```python
df['timestamp'] = df.blockNumber.map(lambda bn: w3.eth.getBlock(bn).timestamp)

df
```

```python
df['date'] = pd.to_datetime(df.timestamp, unit='s')
df['week'] = df.date.map(lambda dt: dt.strftime('%Y-%W'))
```

```python
from matplotlib import pyplot as plt
from datetime import datetime, date, timedelta

plt.rcParams['figure.dpi'] = 200

df['week'] = df.date.map(lambda dt: date.fromisocalendar(dt.isocalendar()[0], dt.isocalendar()[1], 1))

signups = df.groupby('week').size()
ax = signups.plot.line(title="PoH Signups by Week")

ax.set_ylabel('Signups')
ax.set_xlabel('Week')

signups
```
