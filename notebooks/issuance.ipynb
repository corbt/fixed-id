```python
import pandas as pd

# Start from a flat base for simplicity
annual_issuance = [1000 for x in range(0, 70)]

for year in range(1000):
  # Set issuance such that the amount issued in the last 30 years is exactly the same as the amount issued
  # throughout history up through year n-31

  issued_last_29 = sum(annual_issuance[-29:])
  issued_before_last_29 = sum(annual_issuance) - issued_last_29
  issue_this_year = issued_before_last_29 - issued_last_29
  # print([issued_last_29, issued_before_last_29, issue_this_year, annual_issuance[-1]])
  annual_issuance.append(issue_this_year)

pd.Series(annual_issuance).plot.line()
annual_issuance

# list(range(0, 10))[-3:]
```
